### Row 뽑느냐 Column 뽑느냐 문제이다.
- 아래는 한 열을 뽑는 것이다.

### Dictionary or list

### loc 붙이면 row 안붙이면 column

### Indexing
- pandas 에서 여러개의 인자를 가져오고 싶으면 list로 넣어라

```python
df.loc[[0,1,5]]
```

### slicing
- column은 slicing이 안 돼. dictionary 이기 때문이다.
- 엑셀로 예시를 두자.

```python
df.loc[1:3]
```

### 2차원 slicing

```python
df.loc[1:3,['title','rating']]
```

### dictionary 방식과 attribute 방식
- column만 가져올 수 있다.

### loc 방식
- row도 가져올 수 있고 attribute도 가져올 수 있다.

### slicing에서 앞에 껄 생략하면 처음부터, 뒤에를 생략하면 끝까지.
### 양쪽을 생략하면 전부다 나온다.

```python
df.loc[999:,['title', 'rating']]
```

### loc는 label base이다. at은 무조건 하나만 가져옴

```python
df.at[1,'rating']
```

```python
df.loc[1,'rating']
```

### at이 더 빨라

```python
%timeit df.loc[1,'rating']
```

```python
%timeit df.at[1,'rating']
```

### iat
- i 붙은 애들은 index이다.
- 정수값!

at은 label이기 때문에 df.at[1,1] 형식으로 할 수가 없다.

iat은 index방식이기 때문에 df.at[1,1] 가능

```python
df.at[1,1]
```

```python
df.iat[1,1]
```

### 속도는 at이 약간 빠르지만, iat이 편해서 많이 사용한다.

```python
%timeit df.iat[1,1]
```

```python
%timeit df.at[1,'title']
```

### iloc
- python에서 list의 indexing이랑 비슷하다.
- 마지막 column은 나오지 않는다.

### loc는 label base이기 때문에 이름까지 써준다.
- 혼동 주의!

```python
df.loc[1:3,'rating']
```

```python
df.iloc[1:3,1]
```
