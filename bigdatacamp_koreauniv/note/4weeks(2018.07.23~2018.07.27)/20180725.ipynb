### index_column

### Choose only a subset of columns to be read
- 가져올 때부터 버리니?
- 바보 vs 전문가

```python
df = pd.read_csv('data-netflix.csv',usecols=['title','rating','release year'])
df.head()
```

### Handling Missing and NA data
- null 값을 지정해줄 수 있음

```python
df = pd.read_csv('data-netflix.csv',na_values=[''])
```

### Choose whether to skip over blank rows

```python
df = pd.read_csv('data-netflix.csv',skip_blank_lines=False)
```

```python
df.head()
```

### Skip rows

```python
df = pd.read_csv('data-netflix.csv',skiprows=[1,3,7])
```

```python
df.head()
```

### Skip rows from footer or from end of the file

```python
df.tail(2)
df = pd.read_csv('data-netflix.csv',skipfooter=2, engine="python")
df.tail(2)
```

### Reading only a subset of the file or a certain rows

```python
df = pd.read_csv('data-netflix.csv',nrows=100)
# numpy에서 shape많이 씀
# pandas는 numpy 기반.
# python 은 c보다 5~100배 느리다.
df.shape
```

### Reading SQLite3
