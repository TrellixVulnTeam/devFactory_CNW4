## urllib vs requests
### 1. urllib 방식

```python
from urllib.request import Request
from urllib.error import HTTPError

def download(url, agent="python bot", num_retries=2):
    headers = {"User-agent":agent}
    req = Request(url, headers=headers)
    try:
        resp = urlopen(req)
    except HTTPError as e:
        print(e.code, e.reason, e.headers)
        if 500<=e.code<600 and num_retries>0:
            return download(url, num_retries=num_retries-1)

    
    return resp
```

### 헤더 속여서 가져오는 방법!
- agent를 개발자 도구에서 가져온 다음,
- html = download(" ",agent=agent)

```python
agent = "Mozilla/5.0 (iPad; CPU OS 11_0 like Mac OS X) AppleWebKit/604.1.34 (KHTML, like Gecko) Version/11.0 Mobile/15A5341f Safari/604.1"
html = download("https://m.search.naver.com/search.naver?ie=UTF-8&query=%ED%95%9C%EA%B8%80&sm=chr_hty&where=m",agent=agent)
```

```python
html.read().decode("utf-8")
```

### 2. requests 방식

```python
import requests

def download(url, agent="python bot", num_retries=2):
    headers = {"User-agent":agent}
    resp = requests.request("get",url, headers=headers)
    
    if 500 <= resp.status_code < 600 and num_retries > 0:
        print(resp.status_code, resp.reason)
        return download(url, num_retries=num_retries-1)
    
    return resp
```

```python
agent = "Mozilla/5.0 (iPad; CPU OS 11_0 like Mac OS X) AppleWebKit/604.1.34 (KHTML, like Gecko) Version/11.0 Mobile/15A5341f Safari/604.1"
html = download("https://m.search.naver.com/search.naver?ie=UTF-8&query=%ED%95%9C%EA%B8%80&sm=chr_hty&where=m",agent=agent)
```

```python
html.text
```

```python
html.encoding
```

### post 방식으로 보내기
- Json 사용

```python
import requests

def downloadJson(url, agent="python bot", num_retries=2):
    headers = {"User-agent":agent}
    resp = requests.request("post",url, headers=headers, json=jsonParams)
    
    if 500 <= resp.status_code < 600 and num_retries > 0:
        return download(url, num_retries=num_retries-1)
    
    return resp
```

```python
html = downloadJson("http://httpbin.org/post")
```

```python
print(html.text)
```

### get 방식으로 보내기

```python
def downloadGet(url, agent="python bot", num_retries=2):
    headers = {"User-agent":agent}
    resp = requests.get(url, headers=headers)
    
    if 500 <= resp.status_code < 600 and num_retries > 0:
        print(resp.status_code, resp.reason)
        return download(url, num_retries=num_retries-1)
    
    return resp
```

```python
parents = node.find_parents()
type(parents)
```

```python
children = parent.find_all(recursive=False)
type(children)
```

```python
prevsibling = children[1].find_previous_sibling()
prevsibling
```
