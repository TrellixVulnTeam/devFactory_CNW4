### 파일을 열면 항상 닫아줘야돼.
- 하지만 파이썬에서는 알아서 없애줌. 가비지컬렉션 개념

하지만 너무 많이 한번에 열면 에러가 남
- 그래서 쓰는게 .close()

```python
files = []
for i in range(10000):
    files.append(open('moon.txt').close())
```

### with 구문
- 자동으로 마지막에는 exit을 해줘
- 똑같은것 : try - finally
- 하지만 with가 조금 더 세련됨

```python
with open('moon.txt') as f:
    print(f.read())
```
