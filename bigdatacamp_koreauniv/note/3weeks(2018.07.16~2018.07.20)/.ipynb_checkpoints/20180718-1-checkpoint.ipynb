### 왜 byte가 중요한가?
- encoding된 텍스트가 전송되는 것이 아니다.
- 통신도 byte로 전송되기 때문이다.

유니코드
- 전 세계에 있는 모든 문자열의 uplus 방식의 codepoint로 표준화한 것
- 특정 방식으로 저장하기 위해 utf-8도 쓰고 euc-kr 등으로 표현

인코딩
- 암호화 시켜줌
- 사람중심

디코딩
- 인코딩 된 것을 다시 byte로 변환시켜주는 것
- 즉 표준화된 형식으로 다시 변환.
- python3에서는 unicode가 표준
- python2에서는 byte가 표준

byte는 파일을 저장을 하거나 통신상 이용하는 기본 포맷, 기계를 위한 포맷

인코딩 된 것은 사람을 위한 포맷!

----

byte는 8비트, base64는 7비트. 
- 이미지도 텍스트 형식으로 저장할 수 있다.
- 메일 속 첨부되는 사진도 다 base64형식.

python3는 기본적으로 unicode 체계이다.
str - unicode
byte, bytearray - byte

unicode 
- 어떤 방식으로 문자열을 표현할 것이야.
- sys.stdin.encoding...등으로 확인가능

encoding
- 특정한 방식이기 때문에 인코딩 마다 표현할 수 있는 문자가 조금씩 다름

decoding
- 인코딩 된 것을 표준화방식으로 바꾸는 것

base64
- ASCII보다 더 작은 방식으로 인코딩한다.
